```python
df.drop(columns=['id'],inplace=True)
x=df.iloc[:,1:]
y=df.iloc[:,0]
from sklearn.model_selection import train_test_split
```

```python
x
```

```python
y
```

```python
# train test split 

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=90)
```

```python
x_train
```

```python

from sklearn.decomposition import PCA

extraction=PCA(n_components=5)
x_train_tr=extraction.fit_transform(x_train)
x_test_tr=extraction.transform(x_test)
```

```python
principal_components = extraction.components_

# Get the column names of your original dataset
feature_names = list(x.columns)

# Print the feature names that contribute the most to each principal component
for i, component in enumerate(principal_components):
    indices = component.argsort()[::-1]
    top_feature_names = [feature_names[idx] for idx in indices]
    print(f"Principal Component {i+1} top features: {', '.join(top_feature_names)}")
```

```python
x_train_tr.shape
```

```python

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train_tr)
x_test=scaler.transform(x_test_tr)
```

```python
x_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_KNN=KNeighborsClassifier(n_neighbors=10,weights='distance',p=1)

model_KNN.fit(x_train,y_train)
```

```python
knn_pred=model_KNN.predict(x_test)
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score

print(f"the accuracy of KNN model is {accuracy_score(y_test,knn_pred)}")
print(f"the confusion matrix of KNN model is {confusion_matrix(y_test,knn_pred)}")
pos_label = 'M' 
precision = precision_score(y_test, knn_pred, pos_label=pos_label)
print(f"The precision score of the KNN model is {precision}")
precision = recall_score(y_test, knn_pred, pos_label=pos_label)
print(f"The recall score of the KNN model is {precision}")
print(f"the  f1_score of KNN model is {f1_score(y_test,knn_pred, pos_label=pos_label)}")
```

```python
from sklearn.linear_model import LogisticRegression
model_lr=LogisticRegression()

model_lr.fit(x_train,y_train)
```

```python
lr_pred=model_lr.predict(x_test)
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score

print(f"the accuracy of Logistic Regression model is {accuracy_score(y_test,lr_pred)}")
print(f"the confusion matrix of Logistic Regression model is {confusion_matrix(y_test,lr_pred)}")
pos_label = 'M' 
precision = precision_score(y_test, lr_pred, pos_label=pos_label)
print(f"The precision score of the Logistic Regression model is {precision}")
precision = recall_score(y_test, lr_pred, pos_label=pos_label)
print(f"The recall score of the Logistic Regression model is {precision}")
print(f"the  f1_score of Logistic Regression model is {f1_score(y_test,lr_pred, pos_label=pos_label)}")
```
